# Plot Accelerometer Data

```python
df = df.sort_values(by='time', ascending=True)

fig = go.Figure()
fig.add_trace(go.Scatter(x=df["time"], y=df['x'], mode='lines', name = "X axis"))
fig.add_trace(go.Scatter(x=df["time"], y=df['y'], mode='lines', name = "Y axis"))
fig.add_trace(go.Scatter(x=df["time"], y=df['z'], mode='lines', name = 'Z axis'))

fig.update_layout({
'plot_bgcolor': 'rgba(0, 0, 0, 0)',
'paper_bgcolor': 'rgba(0, 0, 0, 0)',
})


fig.update_layout(title='Plot Title',xaxis_title="X Axis Title", yaxis_title="Y Axis Title",height=600, width=1000)
fig.show()
```
